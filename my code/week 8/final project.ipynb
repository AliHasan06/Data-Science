```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import datetime

# Load the data
df = pd.read_csv(r'D:\Data-Science\Dataset\Car details v3.csv')

# Preprocessing steps for each column
current_year = datetime.datetime.now().year
df['car_age'] = current_year - df['year']
df = df.drop('year', axis=1)

df['Make'] = df['name'].str.split().str[0]
df['Model'] = df['name'].str.split().str[1]
make_encoder = LabelEncoder()
model_encoder = LabelEncoder()
df['Make'] = make_encoder.fit_transform(df['Make'])
df['Model'] = model_encoder.fit_transform(df['Model'])
df = df.drop('name', axis=1)

df = pd.get_dummies(df, columns=['fuel'], drop_first=True)
df['transmission'] = df['transmission'].map({'Manual': 0, 'Automatic': 1})

owner_mapping = {'First Owner': 1, 'Second Owner': 2, 'Third Owner': 3, 'Fourth & Above Owner': 4, 'Test Drive Car': 0}
df['owner'] = df['owner'].map(owner_mapping)

df['mileage'] = df['mileage'].str.extract(r'(\d+.\d+)').astype(float)
df['engine'] = df['engine'].str.extract(r'(\d+)').astype(float)
df['max_power'] = df['max_power'].str.extract(r'(\d+.\d+)').astype(float)

# Refined code for extracting numeric values from 'torque' column
df['torque'] = df['torque'].str.extract(r'(\d+\.?\d*)').astype(float)

# Convert 'seats' to float if it's not already numeric
df['seats'] = df['seats'].astype(float)

# Display the preprocessed DataFrame
print(df.head())
```
