# matplotlib

```python
# Importing Matplotlib
# Import the `pyplot` module from matplotlib, which provides a convenient interface for plotting. It is conventionally imported as `plt`.
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.figure(figsize=(10, 5))

plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
y2 = np.cos(x)
y2
```

```python
plt.figure(figsize=(10, 5))
```

```python
plt.figure(figsize=(10, 5))
y2 = np.cos(x)
plt.plot(x, y, label='Sine Wave', color='blue')
plt.plot(x, y2, label='Cosine Wave', color='red', linestyle='--')  # Dashed line
plt.title("Sine and Cosine Waves")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()  # Adding a legend
plt.grid(True)
plt.show()
```
