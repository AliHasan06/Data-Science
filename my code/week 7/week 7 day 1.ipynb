```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

```

# underfitting 

```python
# Step 1: Generate a non-linear dataset
np.random.seed(0)
X = np.sort(5 * np.random.rand(80, 1), axis=0)
y = np.sin(X).ravel() + np.random.normal(0, 0.1, X.shape[0])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# lets use another model
from sklearn.ensemble import RandomForestRegressor

# Creating a Random Forest Regressor model and training it
rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)

# Making predictions
y_pred = rf_model.predict(X_test)

# Evaluating the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
print("Predictions for Test Set:", y_pred)
```

```python
# Step 4: Plot the training data, the true function, and the model's predictions
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.scatter(X_test, y_test, color='green', label='Testing Data')
plt.plot(X, np.sin(X), color='red', label='True Function')
plt.plot(X, rf_model.predict(X), color='orange', label='Model Prediction')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Underfitting Example')
plt.legend()
plt.show()
```

# overfitted



```python
# Example: Overfitting with Decision Tree
from sklearn.tree import DecisionTreeRegressor

# Fit a Decision Tree model
model = DecisionTreeRegressor(max_depth=11)  # High depth to illustrate overfitting
model.fit(X, y)
```

```python
# Plotting the results
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, model.predict(X), color='red', label='Overfitted Tree')
plt.title('Example of Overfitting')
plt.legend()
plt.show()
```
